```python
import pandas as pd

# Carregar o arquivo CSV enviado pelo usurio
file_path = "/mnt/Aviation safety.csv"
df = pd.read_csv(file_path)

# Exibir as primeiras linhas para entender a estrutura dos dados
df.head()
```

```python
import matplotlib.pyplot as plt

# Selecionar as 10 primeiras companhias areas para visualizao
subset = df.head(10)

# Calcular o total de fatalidades por perodo
airlines = subset['airline']
fatalities_85_99 = subset['fatalities_85_99']
fatalities_00_14 = subset['fatalities_00_14']

# Criar grfico de barras
x = range(len(airlines))
width = 0.35

plt.figure(figsize=(12, 6))
plt.bar(x, fatalities_85_99, width=width, label='19851999', color='tomato')
plt.bar([i + width for i in x], fatalities_00_14, width=width, label='20002014', color='steelblue')

# Personalizao
plt.xlabel('Companhias Areas')
plt.ylabel('Nmero de Fatalidades')
plt.title('Comparao de Fatalidades por Companhia Area (19851999 vs 20002014)')
plt.xticks([i + width / 2 for i in x], airlines, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
